Dot Products

```python
import numpy as np

a = np.array([1, 3, 5])
b = np.array([2, 4, 6])

# Dot product
dot = np.dot(a, b)
print("Dot product:", dot)
```

Duality(Linear Functional on V)

```python
def T(v):
    a, b = 4, -2  # example coefficients
    return a * v[0] + b * v[1]

v = np.array([3, 5])
print("T(v):", T(v))  # Output: 4*3 + (-2)*5 = 2
```

Duality(Dual Basis)

```python
# Standard basis: e1 = [1,0], e2 = [0,1]
def T1(v):
    return v[0]  # Extracts first coordinate

def T2(v):
    return v[1]  # Extracts second coordinate

e1 = np.array([1,0])
e2 = np.array([0,1])
print("T1(e1):", T1(e1))  # Output: 1
print("T2(e1):", T2(e1))  # Output: 0
print("T1(e2):", T1(e2))  # Output: 0
print("T2(e2):", T2(e2))  # Output: 1
```

Duality(Dual Maps)

```python
# Suppose L is a matrix from R^2 to R^2
L = np.array([[1, 2],
              [0, 1]])

def dual_map(f):
    # f is a linear functional on W=R^2
    def new_functional(v):
        # First apply L to v, then apply f
        Lv = L @ v
        return f(Lv)
    return new_functional

# Example: f(w) = w[0] - w[1]
def f(w):
    return w[0] - w[1]

g = dual_map(f)
print("Dual map applied:", g(np.array([2,3])))  # result: f(L @ [2,3])
```
