Eigen Vectors and Eigen Values for 2d matrix

```python
import numpy as np
A=np.array([[1,2],[3,4]])
eigen_values,eigen_vectors=np.linalg.eig(A)
print("Eigen Values:- ",eigen_values)
print("Eigen Vectors:- ",eigen_vectors)
```

Eigen Vectors and Eigen Values for 3d matrix

```python
import numpy as np
A=np.array([[1,2,5],[3,4,6],[7,8,9]])
eigen_values,eigen_vectors=np.linalg.eig(A)
print("Eigen Values:- ",eigen_values)
print("Eigen Vectors:- ",eigen_vectors)
```
