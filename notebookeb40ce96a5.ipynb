**Change of Basis**

```python
import numpy as np
#vector in alice basis
alice_v=np.array([2,1])

#bob basis vectors in columns
bob_basis=np.array([[0.8,-1.3],[1.5,0.3]])

#coordinates of alice vector in bob basis
bob_basis_coord=np.linalg.inv(bob_basis) @ alice_v
print("Coordinates of Alice's vector in bob basis:- ",bob_basis_coord)

#Reconstruct alice vector from coordinate of bob basis
alice_v_check= bob_basis @ bob_basis_coord
print("Verifying alice vector based on coordinates:- ",alice_v_check)
```
