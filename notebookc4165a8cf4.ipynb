Co-variance matrix and PCA 

```python
import numpy as np
size_x1=np.array([5,3,10,2])
no_of_rooms_x2=np.array([10,7,15,4])

```

Step-1:- Calculate mean of every feature 

```python
x1_mean=size_x1.mean()
print(x1_mean)
x2_mean=no_of_rooms_x2.mean()
print(x2_mean)
```

Step-2:- Calculate Co-Variance Matrix

```python
m=np.array([size_x1,no_of_rooms_x2])
m_cov=np.cov(m)
print(m_cov)
```

Step-3:- Calculate Eugen Values and Eigen Vectors

```python
m_eigen_value,m_eigen_vectors=np.linalg.eig(m_cov)
print(m_eigen_value,m_eigen_vectors)
```

```python
ev=m_eigen_vectors[1].reshape(2,1)
ev
```

```python
m_transpose=m.transpose()
m_transpose
```

```python
result=np.dot(m_transpose,ev)
result
```

```python
from sklearn.datasets import load_iris
import numpy as np

#Step-1:- Load dataset
data=load_iris()
x=data.data
print("\nInitial data of (150Ã—4) size:- \n",x[:10])
y=data.target

#Step-2:- Calculate Mean
mean=np.mean(x,axis=0)
print("\nMean: \n",mean)

#Step-3:- Subtract Mean from x 
x1=x-mean

#Step-4:- Calculate Covariance
cov=np.cov(x1.T)
print("\nCovariance:- \n",cov)

#Step-5:- Calculate Eigen Values and Eigen Vectors
eigen_values,eigen_vectors=np.linalg.eig(cov)
print("\nEigen Values: \n",eigen_values)
print("\nEigen Vectors: \n",eigen_vectors[:10])

#Step-6:- Sort eigen vectors in Descending Order
eigen_vectors=eigen_vectors.T
indices=np.argsort(eigen_values)[::-1] #output is [3 2 1 0]
print("\nIndices: \n",indices)
eigen_values=eigen_values[indices]
eigen_vectors=eigen_vectors[indices]
print("\nEigen Values: \n",eigen_values)
print("\nEigen Vectors: \n",eigen_vectors[:10])


#Step-7:- Choose First 2 eigen vectors as Components that have high variance
components=eigen_vectors[:2]

#Step-8:- Projections with dot product
results=np.dot(x1,components.T)
print("\nFinal Results after doing PCA of (150Ã—2) size:- \n",results)
```
